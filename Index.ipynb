# Контекстно свободный язык произведения линейных многочленов от «x» с коэффициентами, записанными конечными десятичными дробями 

Обозначения нетерминальных символов:

    < multi_poly > - формула произведения
    < lin_poly >  - линейный полином
    < continuation > - продолжение
    < operand > - операция (сложения/вычитания)

Пояснение: умножение как таковое не было включено в нетерминальный символ <operand> по той причине, что оно является связкой между полиномами и вариативность типа сложение/вычитание/умножение не предусмотрена в качестве связки полиномов 


# Правила грамматики:
    
    <multi_poly> ::= <lin_poly><continuation>;
    
    < continuation > ::= < * >< multi_poly > | < Λ >;
    
    <lin_poly> ::= <(><operand>  <const> <x> <operand> <const> <)> | <(> <const> <x> <operand> <const> <)> | 
    <(><operand>  <const> <operand> <const> <x> <)> | 
    <(> <x> <operand> <const> <)>  | 
    <(> <const> <operand> <x>  <)> | 
      <const><x> | <x> | <operand> <const> | <const> ;
      
    < const > :: = константа \in\ ℝ/(отриц. и 0); - терминальный символ
    
    < operand >::= + | - ;
    
    <*> ::= *; -терминальный символ
    
    < x >::=x; - терминальный символ
    
    <(>::= (; - терминальный символ
    
    <)>::=); - терминальный символ
    




## Грамматический разбор представленного в задании выражения:
    
    2*(x-3,7)*(0,54x+1)*(-0,23+25x)= <const> <*> <(> <x> <operand> <const> <)> <*> <(> <const> <x> <operand> <1> <)> <*> <(> <operand> <const> <operand> <const> <x> <)> < Λ > = <lin_poly> <*> <lin_poly> <*> <lin_poly> <*> <lin_poly> < continuation > = <lin_poly> <*> <lin_poly> <*> <lin_poly> <*> <multy_poly> = <lin_poly> <*> <lin_poly> <*> <lin_poly>< continuation > = <lin_poly> <*> <lin_poly> <*> <multy_poly>=<lin_poly> <*> <lin_poly>< continuation > = <lin_poly> <*> <multy_poly> = <lin_poly> <continuation> = <multy_poly>

Вывод: утверждение верно

##Проверка на однозначность ветвления:
Докажем, что представленная выше грамматика обладает свойством однозначности ветвления по первому символу:

	1)Первое правило выполняется, т.к. независимо от расшифровки <lin_poly> мы переходим в нетерминальный символ <continuation>, который даёт однозначность ветвления: уйти в рекурсию или закончить выражение.
	2)При расшифровке < multy_poly > видим, что <lin_poly> не содержит в себе символ пустоты, а значит нет необходимости перескакивать этот шаг, что не позволяет анализатору содержать пустые ветвления (кроме завершения выражения).

Таблицы правил-диаграмм синтаксического анализатора [ссылка] (https://docs.google.com/document/d/1ka3wwjAI3qf1aQCUqBLeqtN9ZhDfNcxphKeiHurn_4Q/edit?usp=sharing)

```python
import re

def MULPOLY(expr):
    count = 0
    p1 = re.compile(r"^[+-]?\d+(,\d+)?x?$")
    p2 = re.compile(r"^[+-]?x$")
    p3 = re.compile(r"^\([+-]?\d+(,\d+)?x[+-]\d+(,\d+)?\)$")
    p4 = re.compile(r"^\([+-]?x[+-]\d+(,\d+)?\)$")
    p5 = re.compile(r"^\([+-]?\d+(,\d+)?[+-]\d+(,\d+)?x\)$")
    p6 = re.compile(r"^\([+-]?\d+(,\d+)?[+-]x\)$")
    expr = expr.split('*')
    print(expr)
    for ch in expr:
        if p1.search(ch) or p2.search(ch) or p3.search(ch) or p4.search(ch) or p5.search(ch) or p6.search(ch):
            count+=1
    if len(expr) == count:
        print("строка распознанна")
    else:
        print("новозможно распознать строку")
    
expr = str(input())
rt = expr.split('*')
MULPOLY(expr)
```
